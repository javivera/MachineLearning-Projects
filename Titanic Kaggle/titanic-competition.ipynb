```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

combine = [train, test]
```

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

pd.crosstab(train['Title'], train['Sex'])
```

```python
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    
train[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
median = train["Age"].median()
train["Age"].fillna(median, inplace=True)
test["Age"].fillna(median, inplace=True)

for dataset in combine:    
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age']
    
    dataset['Age'] = dataset['Age'].astype(int)
```

```python
for dataset in combine:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1

train[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 1, 'IsAlone'] = 1

train[['IsAlone', 'Survived']].groupby(['IsAlone'], as_index=False).mean()
```

```python
# Lets drop usless features

y_train = train["Survived"]
train = train.drop(['Parch', 'SibSp', 'FamilySize','Name','PassengerId','Cabin','Ticket','Survived'], axis=1)
test = test.drop(['Parch', 'SibSp', 'FamilySize','Name','PassengerId','Cabin','Ticket'], axis=1)

combine = [train,test]
```

```python
train.info()
```

```python
# Lets complete the data

freq_port = train.Embarked.dropna().mode()[0]

for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port)
```

```python
test['Fare'].fillna(test['Fare'].dropna().median(), inplace=True)
```

```python
test.head()
```

```python
for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare']
    dataset['Fare'] = dataset['Fare'].astype(int)
```

```python
# Lets create a simple imputer for the Age 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
    

title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
    
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)

for dataset in combine:
    dataset['Age*Class'] = dataset.Age * dataset.Pclass
```

## Train a model and make prediction

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import cross_val_score
```

```python
X_train = train
X_test = test
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
Y_pred = log_reg.predict(X_test)
acc_log_reg= (sum(cross_val_score(log_reg, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_log_reg
```

```python
coeff_df = pd.DataFrame(train.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(log_reg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
# Support Vector Machines

svc = SVC()
svc.fit(X_train, y_train)
Y_pred = svc.predict(X_test)
acc_svc = acc_perceptron = (sum(cross_val_score(svc, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_svc
```

```python
#Knn

knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
Y_pred = knn.predict(X_test)
acc_knn =  (sum(cross_val_score(knn, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_knn
```

```python
# Perceptron

perceptron = Perceptron()
perceptron.fit(X_train, y_train)
Y_pred = perceptron.predict(X_test)

acc_perceptron = acc_perceptron = (sum(cross_val_score(perceptron, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_perceptron
```

```python
# Linear SVC

linear_svc = LinearSVC()
linear_svc.fit(X_train, y_train)
Y_pred = linear_svc.predict(X_test)

acc_linear_svc = acc_perceptron = (sum(cross_val_score(linear_svc, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_linear_svc
```

```python
sgd = SGDClassifier()
sgd.fit(X_train, y_train)
y_pred = sgd.predict(X_test)

acc_sgd = acc_perceptron = (sum(cross_val_score(sgd, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_sgd
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
Y_pred = decision_tree.predict(X_test)

acc_decision_tree = (sum(cross_val_score(decision_tree, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_decision_tree
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
Y_pred = random_forest.predict(X_test)

acc_random_forest = (sum(cross_val_score(random_forest, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_random_forest
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Perceptron', 
              'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log_reg, 
              acc_random_forest, acc_perceptron, 
              acc_sgd, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
{'criterion': ['gini', 'entropy']},
{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6]}
]

grid_search = GridSearchCV(random_forest, param_grid)
grid_search.fit(X_train,y_train)

acc_random_forest = (sum(cross_val_score(grid_search, X_train, y_train, cv=5, scoring="accuracy")) /5)*100
acc_random_forest
```

```python
grid_search.best_estimator_
```
